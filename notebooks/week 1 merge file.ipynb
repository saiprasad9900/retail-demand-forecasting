```python


calendar = pd.read_csv("../data/processed/calendar_cleaned.csv.gz")

sell_prices = pd.read_csv(
    "../data/processed/sell_prices_cleaned.csv.gz",
    nrows=100000
)

merged_df = pd.merge(
    calendar,
    sell_prices,
    on="wm_yr_wk",
    how="left"
)

merged_df.to_csv(
    "../data/processed/calendar_sell_prices_merged_sample.csv.gz",
    index=False,
    compression="gzip"
)

print("Merged Shape:", merged_df.shape)
```
