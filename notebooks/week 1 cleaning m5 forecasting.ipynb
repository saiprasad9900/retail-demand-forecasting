## Data Cleaning

The following cleaning steps were performed:
- Standardized column names.
- Removed duplicate records.
- Removed extra spaces from text columns.
- Converted date columns into proper datetime format.

```python
def clean_data(df):
    # Standardize column names
    df.columns = df.columns.str.lower().str.strip()

    # Remove duplicate rows
    df = df.drop_duplicates()

    # Remove extra spaces from text columns
    text_columns = df.select_dtypes(include="object").columns
    for col in text_columns:
        df[col] = df[col].str.strip()

    return df
```

```python
calendar = clean_data(calendar)
sales = clean_data(sales)
sales_val = clean_data(sales_val)
sell_prices = clean_data(sell_prices)
sample = clean_data(sample)
```

## add usefull Date Features

```python
# Convert date column to datetime format
calendar['date'] = pd.to_datetime(calendar['date'])

# Create useful date features
calendar['year'] = calendar['date'].dt.year
calendar['month'] = calendar['date'].dt.month
calendar['week'] = calendar['date'].dt.isocalendar().week.astype(int)
calendar['day'] = calendar['date'].dt.day
calendar['day_of_week'] = calendar['date'].dt.day_name()
calendar['day_of_year'] = calendar['date'].dt.dayofyear
calendar['quarter'] = calendar['date'].dt.quarter
calendar['is_weekend'] = calendar['date'].dt.dayofweek >= 5

# Check new features
calendar.head()
```

## Feature Engineering

Created new date-based features from the `date` column to support time-series analysis and identify seasonal patterns.
